```python
#Desenvolva um programa que peça dois números ao usuário e mostre a soma entre eles.
#input() - faz leitura de uma variavel em texto
#int(input()) - converter para inteiro
#float (input()) - converter para decimal

numA = int(input("digite o primeiro número"))
numB = int(input("digite o segundo número"))
print(f"{numA} + {numB} = {numA + numB}")
```

```python
#Crie um programa que peça um número inteiro e mostre seu antecessor e seu sucessor.

numero = int(input("digite o numero: "))
print(f"antecessor: {numero-1} / sucessor:{numero+1}")


```

```python
#Faça um programa que peça o ano de nascimento do usuário e calcule sua idade aproximada.

ano = int(input("digite o ano seu an de nascimento"))
print(f"a idade do usuario e de aproximadamente: {2026-ano}")



```

```python
#Desenvolva um programa que peça duas notas e calcule a média entre elas.

notaA = float(input("digite a primeira nota: "))
notaB = float(input("digite a segunda nota: "))
print(f"a media do aluno é: {(notaA+notaB)/2}")


```

```python
#Crie um programa que peça uma temperatura em graus Celsius e converta para Fahrenheit.

tempC = float(input("digite a temperatura: "))
print(f"a temperatura em fahrenheit é: {(tempC*9/5)+32}")
```

```python
#Faça um programa que peça a largura e a altura de um retângulo e calcule sua área.
#:.2f - duas casas após avírgula

altura = float(input("digite a altura"))
largura = float(input("digite a largura"))
print (f"a área do retangulo é de: {(altura*largura)}")
```
