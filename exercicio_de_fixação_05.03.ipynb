```python
#Solicite o preço de um produto e o percentual de desconto, converta ambos para float, calcule o valor final com desconto aplicado e exiba o resultado.

preco = float(input("digite o valor do produto: "))
desconto = float(input("digite o valor do percentual de desconto: "))
vf = (preco - (preco * desconto / 100))

print(f"o valor do seu produto com desconto e de: {vf}")
```
